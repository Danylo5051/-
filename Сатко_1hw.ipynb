```python
import numpy as np
```

```python
vector_20 = np.arange(1, 21)
print("Вектор натуральних чисел:\n", vector_20)
```

```python
ones_matrix = np.ones((4, 4))
print("\nМатриця 4x4 з одиниць:\n", ones_matrix)
```

```python
random_matrix_5x5 = np.random.uniform(-1, 1, (5, 5))
print("Матриця 5x5 (від -1 до 1):\n", random_matrix_5x5)
```

```python
random_matrix_3x3 = np.random.rand(3, 3)
print("\nМатриця 3x3 (від 0 до 1):\n", random_matrix_3x3)
```

```python
arr1= np.random.randint(1, 21, 6)
arr2= np.random.randint(1, 21, 6)

print(f"Масив 1: {arr1}")
print(f"Масив 2: {arr2}")
print("Додавання:", arr1 + arr2)
print("Віднімання:", arr1 - arr2)
print("Множення:", arr1 * arr2)
```

```python
mat_a = np.random.randint(1, 6, (3, 2))
mat_b = np.random.randint(1, 6, (2, 4))
result_dot = np.dot(mat_a, mat_b)

print("Матриця 3x2:\n", mat_a)
print("Матриця 2x4:\n", mat_b)
print("Результат множення (3x4):\n", result_dot)
```

```python
mat_to_transpose = np.random.rand(3, 3)
print("Оригінальна матриця:\n", mat_to_transpose)
print("Транспонована матриця:\n", mat_to_transpose.T)
```

```python
mat_c = np.random.randint(1, 11, (3, 3))
mat_d = np.random.randint(1, 11, (3, 3))
print("\nМатриця C:\n", mat_c)
print("Матриця D:\n", mat_d)
print("Різниця C - D:\n", mat_c - mat_d)
```

```python
mat_e = np.random.randint(-5, 6, (3, 3))
mat_f = np.random.randint(-5, 6, (3, 3))
print("Матриця E:\n", mat_e)
print("Матриця F:\n", mat_f)
print("Матричний добуток E * F:\n", np.dot(mat_e, mat_f))
```

```python
mat_g = np.random.uniform(-1, 1, (3, 3))
row_sums = mat_g.sum(axis=1).reshape(-1, 1)
```

```python
print("\nМатриця G:\n", mat_g)
print("Вектор-стовпчик сум рядків:\n", row_sums)
```
