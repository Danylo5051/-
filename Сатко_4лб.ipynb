```python
df['month'] = pd.to_datetime(df['month'], format='%m/%d/%Y %I:%M:%S %p')

df_2018 = df[df['month'].dt.year == 2018]

min_visitors = df_2018.iloc[:, 1:].min()
max_visitors = df_2018.iloc[:, 1:].max()
result = pd.DataFrame({'min': min_visitors, 'max': max_visitors})

result
```

```python
average_visitors = df.iloc[:, 1:].mean()
average_visitors
```

```python
df_2015 = df.loc[df['month'].dt.year == 2015].copy()

df_2015['total_visitors'] = df_2015.iloc[:, 1:].sum(axis=1)

min_row = df_2015.loc[df_2015['total_visitors'].idxmin()]
max_row = df_2015.loc[df_2015['total_visitors'].idxmax()]

min_month = min_row['month'].strftime('%B')
max_month = max_row['month'].strftime('%B')

print(f'Найнижча відвідуваність у {min_month}')
print(f'Найвища відвідуваність у {max_month}')
```

```python
df_2015.head()
```

```python
df_2018 = df[df['month'].dt.year == 2018].copy()

df_2018['month_num'] = df_2018['month'].dt.month

summer = df_2018[df_2018['month_num'].isin([6, 7, 8])]['avila_adobe'].sum()

winter = df_2018[df_2018['month_num'].isin([1, 2, 12])]['avila_adobe'].sum()

print(f" Літом відвідали: {summer} людей")
print(f" Зимою відвідали: {winter} людей")
```

```python
mtx = df.corr(numeric_only=True).abs()


fig, ax = plt.subplots(figsize=(8, 8))

sns.heatmap(
    mtx,
    cmap='crest',
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    mask=np.triu(np.ones_like(mtx, dtype=bool)),
    square=True,
    cbar=False,
    ax=ax
)

plt.title("Матриця попарної кореляції ознак", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# 1. Лінійний графік
sns.set_theme(style="darkgrid")
sns.lineplot(data=df, x='month', y='avila_adobe')
plt.title('Avila Adobe: Відвідуваність з часом')
plt.show()
```

```python
# 2. Boxplot
df['year'] = df['month'].dt.year
sns.boxplot(data=df, x='year', y='chinese_american_museum')
plt.title('Chinese American Museum: Розподіл по роках')
plt.show()
```

```python
# 3. Histogram
sns.histplot(df['gateway_to_nature_center'], bins=30, kde=True)
plt.title('Розподіл відвідуваності Gateway to Nature Center')
plt.show()
```

```python
# 5. Barplot
avg_visitors = df.drop(columns=['month', 'year']).mean().sort_values(ascending=False)
sns.barplot(x=avg_visitors.values, y=avg_visitors.index)
plt.title('Середня кількість відвідувачів по музеях')
plt.xlabel('Середнє значення')
plt.ylabel('Музей')
plt.show()
```

```python
# Завдання 2. Завантажити датасет "Пінгвіни", вивести дані.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Завантажуємо вбудований датасет
penguins = sns.load_dataset("penguins")

# Виводимо перші рядки
print(penguins.head())
```

```python
# 1. Двовимірний графік щільності (Bivariate KDE Plot)
plt.figure(figsize=(8, 6))
sns.kdeplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species", fill=True, alpha=0.6, palette=['#4a148c', '#9c27b0', '#ce93d8'])
plt.title('Двовимірна щільність: довжина плавця vs маса тіла')
plt.show()
```

```python
# 2. Regplot (Лінійна регресія: залежність маси від довжини плавця)
sns.set_theme(style="darkgrid")
sns.regplot(data=penguins, x='flipper_length_mm', y='body_mass_g', color='darkviolet', scatter_kws={'alpha': 0.5})
plt.title('Регресійна модель: маса тіла vs довжина плавця')
plt.show()
```

```python
# 3. Lmplot (Лінійна регресія для кожної статі окремо)
lm = sns.lmplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="sex", palette=['indigo', 'violet'], height=6, aspect=1.2)
plt.title('Лінійна регресія за статтю', y=1.05)
plt.show()
```

```python
# 4. Pointplot (Оцінка середньої глибини дзьоба за видами та статтю)
sns.set_theme(style="darkgrid")
sns.pointplot(data=penguins, x='species', y='bill_depth_mm', hue='sex', palette=['indigo', 'violet'], dodge=True)
plt.title('Оцінка середньої глибини дзьоба за видами')
plt.show()
```

```python
# 5. Scatterplot + Rugplot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=penguins, x="body_mass_g", y="flipper_length_mm", color="indigo", alpha=0.5)
sns.rugplot(data=penguins, x="body_mass_g", y="flipper_length_mm", color="darkviolet", height=0.05)
plt.title('Діаграма розсіювання з відмітками щільності (Rugplot)')
plt.show()
```
